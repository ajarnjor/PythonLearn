```python
import pandas as pd
import requests
import bs4
from bs4 import BeautifulSoup as soup
```

```python
my_url = 'https://www.topuniversities.com/student-info/choosing-university/worlds-top-100-universities'
```

```python
r = requests.get(my_url)
r.status_code
```

```python
r.content
```

```python
s = soup(r.text,"lxml")
type(s)
```

```python
print(s.prettify())
```

```python
#find table
top_Us = s.find('table')
top_Us
```

```python
#extract to list
lst = []
for tr in top_Us.find_all('tr'):
    place = tr.find_all('td')[0].text.strip()
    uni = tr.find_all('td')[1].text.strip()
    country = tr.find_all('td')[2].text.strip()
    print(place,uni,country)
    lst.append([place, uni, country])

lst
```

```python
#Make the dataframe
df = pd.DataFrame(lst, columns = ['Rank','University','Location'])
df.head(5)
```

```python
#delete first rows
df = df.iloc[0:,]
df
```
